```python
% reset

import numpy as np
from scipy import stats

% config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from matplotlib import rcParams
% matplotlib inline
rcParams['figure.figsize'] = 6, 4
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Tahoma']
```

```python
np.random.seed(seed=11111)
bins = 25

samples_gamma = 1500
samples_normal1 = 1750
samples_normal2 = 2250

rand_gamma = np.random.gamma(shape=9.0, scale=0.3, size=samples_gamma)
rand_normal1 = np.random.normal(loc=np.mean(rand_gamma), scale=1.00, size=samples_normal1)
rand_normal2 = np.random.normal(loc=np.mean(rand_gamma), scale=1.01, size=samples_normal2)

plt.hist(rand_normal1, bins=bins, color='orangered', label='normal1', lw=0, alpha=0.5)
plt.hist(rand_gamma, bins=bins, color='deepskyblue', label='gamma', lw=0, alpha=0.5)
plt.legend()
plt.xlim(0, 6)
plt.ylim(0, 300)
plt.show()

plt.hist(rand_normal2, bins=bins, color='forestgreen', label='normal2', lw=0, alpha=0.5)
plt.hist(rand_normal1, bins=bins, color='orangered', label='normal1', lw=0, alpha=0.5)
plt.legend()
plt.xlim(0, 6)
plt.ylim(0, 300)
plt.show()
```

```python
plt.hist(rand_gamma, bins=bins, normed=True, cumulative=True, histtype='step', label='gamma', color='deepskyblue')
plt.hist(rand_normal1, bins=bins, normed=True, cumulative=True, histtype='step', label='normal1', color='orangered')
plt.legend(loc=2)
plt.xlim(0, 6)
plt.show()

plt.hist(rand_normal1, bins=bins, normed=True, cumulative=True, histtype='step', label='normal1', color='orangered')
plt.hist(rand_normal2, bins=bins, normed=True, cumulative=True, histtype='step', label='normal2', color='forestgreen')
plt.legend(loc=2)
plt.xlim(0, 6)
plt.show()
```

```python
d_gn1,p_gn1 = stats.ks_2samp(rand_gamma, rand_normal1)
print(d_gn1, p_gn1)

d_n1n2,p_n1n2 = stats.ks_2samp(rand_normal1, rand_normal2)
print(d_n1n2, p_n1n2)


print('gamma vs. normal1: p = ' + str(p_gn1))
print('normal1 vs. normal2: p = ' + str(p_n1n2))
```
